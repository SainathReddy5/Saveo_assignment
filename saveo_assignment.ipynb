```python
def count(a,b):
    count = 0
    for i in range(len(a)):
        if a[i] == b[i]:
            pass
        else:
            count += 1
        pass

    return count
```

```python
q5_df.status.value_counts(),q5_df.status.isna().sum()
```

```python
def make_df(a,b,c):
    data = []
    for i in range(1,33):
        sum = 0
        ca = 0
        cb = 0
        for j in range(len(a)):
            if a[j] == i:
                sum = sum + b[j]
                pass
            if a[j]==i and c[j] == 'status_a':
                ca = ca + 1
            if a[j]==i and c[j] == 'status_b':
                cb = cb + 1
                pass
            pass
        data.append([i,sum,ca,cb])
    return data    




```

```python
data = make_df(q5_df.distributorId,q5_df.quantity,q5_df.status)
```

```python
data
```

```python
sol5_df = pd.DataFrame(data,columns=['distributorId','totalOrders','status_a(count)','status_b(count)'])
sol5_df
```

```python
bsum = 0
for i in sol5_df['status_b(count)']:
    bsum = bsum + i
bsum    
```

```python
sol5_df['status_b(percentage)'] = sol5_df['status_b(count)'].apply(lambda x:(x/bsum)*100)
```

```python
sol5_df.set_index('distributorId',inplace=True)
```

```python
sol5_df.to_csv('5.csv')
```
